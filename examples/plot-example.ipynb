```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
```

```python
epoch = np.linspace(1, 25)
train_loss = np.random.normal(scale=.01, size=epoch.size) + 1 / epoch + .4
val_loss = np.random.normal(scale=.05, size=epoch.size) + ((.8*epoch - 10)**2)*.01 + 0.7
```

```python
fig, ax = plt.subplots(figsize=(8, 4))
plt.xlim(1, 25)
plt.ylim(0, 2)
plt.minorticks_on()
plt.xlabel("epoch")
plt.ylabel("loss")
plt.title("Training / Validation Loss")
plt.grid(True)
plt.plot(epoch, train_loss, label="Training") 
plt.plot(epoch, val_loss, label="Validation")
plt.legend()
plt.show()
```
